```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [[] for i in range(4)]
inst = [[] for i in range(4)]
dir = {
    "descending": 0,
    "evenorodd": 1,
    "fibonacci": 2,
    "palindrome": 3,
    "prime": 4
}
path = "."
files = os.listdir(path)[:-1]
for file in files:
    with open("./"+file, "r+") as f:
        ipc = float(f.readlines()[-1].split(":")[-1])
        splitted = file.split("_")
        index = int(splitted[1])-1
        out = splitted[-1].split(".")[0]
        if splitted[0]=="data": data[index].append(ipc)
        else: inst[index].append(ipc)
data.append([0.0245, 0.0276, 0.0233, 0.02223, 0.0217])
inst.append([0.0245, 0.0276, 0.0233, 0.02223, 0.0217])
        
```

```python
data_np = np.array(list(reversed(data))).T
inst_np = np.array(list(reversed(inst))).T
data_np
```

```python
bytes = np.array([0, 1, 2, 3, 4])
plt.figure(figsize=(15, 10))
plt.title("IPC vs Data Cache Size (Inst. Cache = 1 KB)", fontsize=20)
plt.xlabel("Data Cache Size (Bytes)", fontsize=20)
plt.ylabel("IPC", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.plot(bytes, data_np[0], label="descending", linewidth=4)
plt.plot(bytes, data_np[3], label="palindrome", linewidth=4)
plt.plot(bytes, data_np[2], label="fibonacci", linewidth=4)
plt.plot(bytes, data_np[4], label="prime", linewidth=4)
plt.plot(bytes, data_np[1], label="evenorodd", linewidth=4)
plt.legend(loc='upper left', fontsize=15)
plt.xticks(bytes, ["0 B", "16 B", "128 B", "512 B", "1024 B"])
plt.savefig("IPCvsData.jpg")
plt.show()
```

```python
bytes = np.array([0, 1, 2, 3, 4])

plt.figure(figsize=(15, 10))
plt.title("IPC vs Inst. Cache Size (Data Cache = 1KB)", fontsize=20)
plt.xlabel("Instruction Cache Size (Bytes)", fontsize=20)
plt.ylabel("IPC", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.plot(bytes, inst_np[0], label="descending", linewidth=4)
plt.plot(bytes, inst_np[3], label="palindrome", linewidth=4)
plt.plot(bytes, inst_np[2], label="fibonacci", linewidth=4)
plt.plot(bytes, inst_np[4], label="prime", linewidth=4)
plt.plot(bytes, inst_np[1], label="evenorodd", linewidth=4)
plt.xticks(bytes, ["0 B", "16 B", "128 B", "512 B", "1024 B"])
plt.legend(loc='upper left', fontsize=15)
plt.savefig("IPCvsInst.jpg")
plt.show()
```
